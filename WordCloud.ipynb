# Palavras mais utilizadas em "A última Pergunta" - Asimov
## Criando uma Word Cloud com Python

Iremos aqui descobrir, vizualizar e contabilizar as palavras (excluindo stop words) mais utilizadas por Isaac Asimov no conto "A última Pergunta".

```python
import collections
import matplotlib.pyplot as plt

file=open('C:/Users/user/Documents/Programação/Python/EDX/word_cloud/A última pergunta - Asimov/a_ultima_pergunta-asimov.txt', encoding="ISO-8859-1")
contador= {}
stopwords = set(line.strip() for line in open('C:/Users/user/Documents/Programação/Python/EDX/word_cloud/A última pergunta - Asimov/stopwords_pt', encoding="ISO-8859-1"))
for word in file.read().lower().split():
    word = word.replace(".","")
    word = word.replace("?","")
    word = word.replace(",","")
    word = word.replace("\"","")
    word = word.replace("“","")
    word = word.replace("\x93","")
    word = word.replace("\x94","")
    word = word.replace("/","")
    if word not in stopwords:
        if word not in contador:
            contador[word] = 1
        else:
            contador[word] += 1
palavras_em_ordem = collections.Counter(contador)

print('As 10 palavras mais utilizadas em "A última pergunta", de Asimov, excluindo uma lista de stopwords são:')
x = {}
for word, count in palavras_em_ordem.most_common(10):
    print(word, ": ", count)
    x[word] = count
f, ax = plt.subplots(figsize=(20,10))
plt.bar(x.keys(), x.values())
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
```

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

file=open('C:/Users/user/Documents/Programação/Python/EDX/word_cloud/A última pergunta - Asimov/a_ultima_pergunta-asimov.txt', encoding="ISO-8859-1")
stopwords = set(line.strip() for line in open('C:/Users/user/Documents/Programação/Python/EDX/word_cloud/A última pergunta - Asimov/stopwords_pt', encoding="ISO-8859-1"))
palavras = ""
for word in file.read().lower().split():
    word = word.replace(".","")
    word = word.replace("?","")
    word = word.replace(",","")
    word = word.replace("\"","")
    word = word.replace("“","")
    word = word.replace("\x93","")
    word = word.replace("\x94","")
    word = word.replace("/","")
    palavras = palavras +' '+ word

wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(palavras)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
```
